**Text Analytics:** Method of Extracting meaning insights and Answering Questions from Text data.

**Processes in Pre-Processing:**
* **Preliminaries:** Sentence Segmentation, Word Tokenization
* **Frequent Steps:** Stopwords Removal, Stemming, Lemmatization, Removing Digits/Punctuation, Lowercasing, etc
* **Other Steps:** Normalization, Language Detection, Code Mixing, Transliteration,etc.
* **Advanced Processing:** POS Tagging, Parsing, co-reference resolution.

**Tokenization:** Process of Splitting Textual data into smaller components called Tokens.

**Different types of Tokenization:**
* **Sentence Tokenization:** Split paragraph into sentence.
* **Word Tokenization:** Split sentences into words.
* **TreebankWord Tokenizer:** Default
* **RegExp Tokenizer:** For Patterns.
* **Whitespace Tokenizer:** Tabs, Newline, Spaces.

**Part-of-speech (POS) tagging:** Assigning grammatical labels (tags) to each word in a sentence such as noun, verb, adjective, adverb, pronoun, etc.

**Uses:**

**Improved Language Understanding:** By understanding the grammatical role of each word, NLP models can better grasp the overall meaning of a sentence.

**Example:**

**spaCy**: This popular NLP library offers efficient POS tagging along with other functionalities like NER and dependency parsing

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk import word_tokenize
```

```python
words = word_tokenize("Iam a Data Science Enthusiast")
```

```python
print(words)
```

```python
print(nltk.pos_tag(words))
```
