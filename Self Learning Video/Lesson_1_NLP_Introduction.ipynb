**Regular Expressions (Regex):**
* Tool for Text Analysis and building Rule-Based Systems.

**Eg: (@[A-Za-z0-9_]+)** finds Twittern Handles.

**Natural Language Understanding:**
* Subset of NLP
* Uses Syntactic and Semantic understanding of text and Speech to determine the meaning of a sentence.,
* Syntax: Grammatical Structure of a sentence
* Semantics: Intended meaning.
* Data Structures that specifies the relationship between words and phrases.

**Natural Language Generation:**
* Subset of NLP.
* Enables computers to write.
* The Process of producing a Human Language text response based on the Input Data.
* NLG systems use templates, hidden Marcov Chain, neural networks,etc
* Uses Language Rules to generate responses.

**Stages of NLG:**
* **Text Planning:** Formulate and Ordering content in a logical manner.
* **Sentence Planning:** Punctuation, Textflow, Paragraphs,sentence breakdowns and add pronouns/conjunctions.
* **Realization:**  Grammatical Accuracy, ensuring rules around punctuation.
