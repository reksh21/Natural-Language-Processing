**Feature extraction:** Helps us identify and extract the most relevant features from the text include things like word frequency, word co-occurrence (how often words appear together), part-of-speech tags, named entities, etc

**Feature:** Piece of Information/Measurable Property.

**Feature Extraction Technique:**
* **Named Entity Recognition(NER):**
  * Identify and classify named entities mentioned in text data.
  * These entities can be various categories of things, such as:
    * People (e.g., Barack Obama, Elon Musk)
    * Organizations (e.g., Apple, Microsoft)
    * Locations (e.g., Paris, Tokyo, Mount Everest)
    * Dates and Times (e.g., August 15, 2024, 2:30 PM)
    * Monetary Values (e.g., $100, €120)
    * Other categories depending on the specific task (e.g., medical codes, product names)
  * **Eg: Stanford NER**

**Examples of Feature Extraction in NLP:**

* **Bag-of-Words (BoW):**
  * for representing text data numerically.
  * It focuses on the frequency of words appearing in a document, but disregards the order or context in which they appear.

image.png

* **Term Frequency-Inverse Document Frequency(TF-IDF):** For weighting words in a document.

  **Term Frequency (TF):** This measures how frequently a word appears in a specific document. It's calculated as the number of times a word appears divided by the total number of words in the document.

  **Inverse Document Frequency (IDF):** This considers how common a word is across all documents in the corpus. Words that appear frequently across all documents (like "the", "a", "an") will have a lower IDF value, reducing their overall weight. Conversely, words that are specific to a particular document will have a higher IDF value, emphasizing their importance.

**TF-IDF = TF * IDF**

* **Word embeddings:**
  * Represent words as vectors in a high-dimensional space.
  * Unlike Bag-of-Words (BoW) and TF-IDF which focus on word frequency, word embeddings capture semantic relationships and meaning between words.
  * Pre-trained Word Embeddings: Word2Vec, GloVe.


* **One-hot encoding:**
  * used to represent categorical data numerically.
  *  It works by creating a new binary vector for each possible category within the data.
  * all elements are zeros except for the position corresponding to the category the data point belongs to, which is set to one.

image.png

**Word2vec:**
* It  represents words as vectors in a high-dimensional space, where these vectors capture the semantic meaning and relationships between words.

**Word2vec uses two main techniques to learn word embeddings:**

**Continuous Bag-of-Words (CBOW):**
The model tries to predict a target word based on its surrounding context words in a sentence. By analyzing how well the model predicts the target word based on its context, the model learns the semantic relationships between words and refines the word vectors.

**Skip-gram:**  Here, the model does the opposite. Given a word (the input word), it tries to predict the surrounding words (context words) within a window. This helps the model understand the meaning of the input word based on the words that typically appear around it.
